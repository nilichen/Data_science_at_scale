```r
train = read.csv("train.csv", stringsAsFactors=FALSE)
```

```r
store = read.csv("store.csv", stringsAsFactors=FALSE)
```

```r
newTrain = merge(train, store, by="Store")
```

```r
head(newTrain)
```

```r
newTrain$Date = as.Date(newTrain$Date)
```

```r
library(dplyr)
salesbydate2 = group_by(newTrain, Promo2, Date) %>% summarise(sales = mean(Sales))
head(salesbydate2)
```

```r
library(ggplot2)
salesbydate2$Promo2 = factor(salesbydate2$Promo2)
ggplot(aes(x=Date, y=sales), data=salesbydate2) + geom_line() + facet_wrap(~Promo2)
```

```r
salesbydate = group_by(newTrain, Promo, Date) %>% summarise(sales = mean(Sales))
salesbydate$Promo = factor(salesbydate$Promo)
ggplot(aes(x=Date, y=sales), data=salesbydate) + geom_line() + facet_wrap(~Promo)
```

```r
dayofweek = group_by(newTrain, DayOfWeek, Date) %>% summarise(sales = mean(Sales))
#dayofweek$DayOfWeek = factor(salesbydate$DayOfWeek)
ggplot(aes(x=Date, y=sales), data=dayofweek) + geom_line() + facet_wrap(~DayOfWeek)
```
